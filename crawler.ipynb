```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
LIST_LINK = []

for i in range(2000):
    try:
        response = requests.get("https://bds.com.vn/mua-ban-nha-dat-ha-noi-page" + str(i + 1))
        soup = BeautifulSoup(response.content, "html.parser")
        list_link = soup.findAll('div', 'item-nhadat')          
        for item in list_link:
            link = item.find('a', 'title-item-nhadat').get('href')
            LIST_LINK.append(
                {
                    'link': link
                }
            )
        print('####Done page' + str(i + 1))
    except Exception as e:
        print('####Error get link ' + str(i + 1))
        print(e)
```

```python
df = pd.DataFrame(LIST_LINK)
df.to_csv('./data/link.csv')
```

```python
df_link= pd.read_csv('data/link.csv')
df_link.info()
```

```python
LIST_LINK = df_link['link'].values.tolist()
```

```python
print(len(LIST_LINK))
```

```python
LIST_BDS = []

for i in range(len(LIST_LINK)):
    try:
        response = requests.get(LIST_LINK[i])
        soup = BeautifulSoup(response.content, "html.parser")

        title = None
        infoAddress = None
        estate_type = None
        province = None
        district = None
        ward = None
        price = None
        square = None
        post_date = None
        describe = None
        numb_bedroom = None
        numb_floor = None

        try:
            title = soup.find('h1', 'title-product').text
        except:
            print('error title')
        try:
            infoAddress = soup.find('ul', 'breadcrumb').findAll('li')
        except: 
            print('error infoAddress')
        try:
            estate_type = ' '.join(infoAddress[0].text.split()[2:])
        except: 
            print('error estate_type')
        try:
            province = infoAddress[1].text
        except:
            print('error province') 
        try:
            district = infoAddress[2].text
        except:
            print('error district') 
        try:
            ward = infoAddress[3].text
        except:
            print('error ward') 
        try:
            important_info = soup.findAll('ul', 'list-attr-hot')[0].findAll('li')
        except:
            print('error important_info') 
        try:
            price = important_info[0].findAll('span')[1].text
        except:
            print('error price') 
        try:
            square = important_info[1].findAll('span')[1].text
        except:
            print('error square') 
        try:
            post_date = important_info[2].findAll('span')[1].text
        except:
            print('error post_date') 
        try:
            describe = soup.find('div', 'ct-pr-sum').text
        except:
            print('error describe') 
        try:
            if len(soup.findAll('ul', 'list-attr-hot')) > 1:  
                extent_info = soup.findAll('ul', 'list-attr-hot')[1].findAll('li')
                for ele in extent_info:
                    if ele.findAll('span')[0].text == 'S phng ng :':
                        numb_bedroom = ele.findAll('span')[1].text
                    elif ele.findAll('span')[0].text == 'Tng :':
                        numb_floor = ele.findAll('span')[1].text
        except:
            print('error extent_info')
        LIST_BDS.append(
            {   'title': title,
                'description': describe,
                'price': price,
                'square': square,
                'estate_type': estate_type,
                'post_date': post_date,
                'link': LIST_LINK[i],
                'province': province,
                'district': district,
                'ward': ward,
                'Tng': numb_floor,
                'numb_bedroom': numb_bedroom,
            }
        )
        print('####Finish get sample ' + str(i + 1))
    except Exception as e:
        print('####Error get sample ' + str(i + 1))
        print(e)
```

```python
print(len(LIST_BDS))
```

```python
df_link_product = pd.DataFrame(LIST_BDS)
df_link_product.to_csv('data/data.csv')
```
